```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Se dau urmatoarele 9 imagini de dimensiuni 400 x 600. Valorile acestora au fost salvate in fisierele "images/car_{id}.npy":
    - a) Cititi imaginile din aceste fisiere si salvatile intr-un np.array
    - b) Calculati suma valorilor pixelilor tuturor imaginilor
    - c) Calculati suma valorilor pixelilor pentru fiecare imagine in parte
    - d) Afisati indexul imaginii cu suma maxima
    - e) Calculati imaginea medie si afisati-o
    - f) Cu ajutorul functiei np.std(images_array), calculati deviatia standard a imaginilor
    - g) Normalizati imaginile. (se scade imaginea medie si se imparte rezultatul la deviatia standard)
    - h) Decupati fiecare imagine, afisand numai liniile cuprinse intre [200, 300], respectiv coloanele cuprinse intre [280, 400].

```python
# a) Cititi imaginile din aceste fisiere si salvatile intr-un np.array;

images = []
for id in range(9):
    img = np.load("images/car_{}.npy".format(id))
    images.append(img)
    
images = np.array(images)

print(np.shape(images))
```

```python
# b) Calculati suma valorilor pixelilor tuturor imaginilor

sumPixelsTotal = np.sum(images)

print(sumPixelsTotal)
```

```python
# c) Calculati suma valorilor pixelilor pentru fiecare imagine in parte

sumPixelsEach = np.sum(images, axis=(1, 2))     # axis: 0 - nrImages, 1 - imgHeight, 2 - imgWidth

print(sumPixelsEach)
```

```python
# d) Aflati indexul imaginii cu suma maxima

maxSumIndex = np.argmax(sumPixelsEach)

print(maxSumIndex)
```

```python
# e) Calculati imaginea medie si afisati-o

meanImg = np.mean(images, axis=(0))

plt.imshow(meanImg, cmap='gray')
plt.show()
```

```python
# f) Cu ajutorul functiei np.std(images), calculati deviatia standard a imaginilor

stDeviation = np.std(images)

print(stDeviation)
```

```python
# g) Normalizati imaginile. (se scade imaginea medie si se imparte rezultatul la deviatia standard)

normalisedImages = (images - meanImg) / stDeviation

for img in normalisedImages:
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
# h) Decupati fiecare imagine, afisand numai liniile cuprinse intre [200, 300], respectiv coloanele cuprinse intre [280, 400].

for img in images:
    croppedImg = img[200:300, 280:400]
    plt.imshow(croppedImg, cmap='gray')
    plt.show()
```
